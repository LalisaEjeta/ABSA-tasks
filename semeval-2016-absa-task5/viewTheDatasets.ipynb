```python
import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.parse("./laptops_train_sb1.xml")
root = tree.getroot()

rows = []

for sentence in root.findall(".//sentence"):
    sent_id = sentence.get("id")
    text = sentence.find("text").text
    
    opinions = sentence.find("Opinions")
    
    if opinions is not None and len(opinions) > 0:
        for op in opinions.findall("Opinion"):
            rows.append({
                "sentence_id": sent_id,
                "text": text,
                "category": op.get("category"),
                "polarity": op.get("polarity")
            })
    else:
        rows.append({
            "sentence_id": sent_id,
            "text": text,
            "category": None,
            "polarity": None
        })

df = pd.DataFrame(rows)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.parse("./laptops_test_sb1.xml")
root = tree.getroot()

rows = []

for review in root.findall(".//Review"):
    review_id = review.get("rid")
    
    for sentence in review.findall(".//sentence"):
        sent_id = sentence.get("id")
        text = sentence.find("text").text
        
        opinions = sentence.find("Opinions")
        
        # if no opinions → still keep row
        if opinions is None or len(opinions) == 0:
            rows.append({
                # "review_id": review_id,
                # "sentence_id": sent_id,
                "text": text,
                "category": None,
                "polarity": None
            })
        else:
            for op in opinions.findall("Opinion"):
                rows.append({
                    # "review_id": review_id,
                    # "sentence_id": sent_id,
                    "text": text,
                    "category": op.get("category"),
                    "polarity": op.get("polarity")
                })

df = pd.DataFrame(rows)

df.head()
```
